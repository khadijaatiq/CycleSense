```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
df = pd.read_csv('../data/synthetic_menstrual_cycle_data.csv')
print("Shape:", df.shape)           # rows, columns
print("\nColumn types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())
```

```python
plt.figure(figsize=(8, 4))
df['cycle_length'].hist(bins=25, color='orchid', edgecolor='white')
plt.title('Distribution of Cycle Lengths')
plt.xlabel('Days')
plt.ylabel('Frequency')
plt.show()
```

```python
avg_by_stress = df.groupby('stress_level')['cycle_length'].mean()
avg_by_stress.plot(kind='bar', color='salmon')
plt.title('Average Cycle Length by Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Avg Days')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
cols = ['cycle_length', 'stress_level', 'sleep_hours',
        'exercise_intensity', 'illness_flag']
sns.heatmap(df[cols].corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
regular_user   = df[df['user_id'] == 1]['cycle_length']
irregular_user = df[df['user_id'] == 5]['cycle_length']

plt.figure(figsize=(10, 4))
plt.plot(regular_user.values,   marker='o', label='Regular user')
plt.plot(irregular_user.values, marker='s', label='Irregular user')
plt.title('Cycle Length Over Time: Regular vs Irregular User')
plt.xlabel('Cycle Number')
plt.ylabel('Days')
plt.legend()
plt.show()
```
